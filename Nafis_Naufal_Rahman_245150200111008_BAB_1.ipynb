```python
umur = int(input("Masukkan umur pengunjug : "))
hari = input("Masukkan hari pembelian tiket : ").lower()

harga_tiket = 50000
diskon = 0
if umur < 12 or umur > 60:
  diskon += 50
elif umur >= 12 and umur <= 60:
  diskon += 0
  if hari in ["senin", "selasa", "rabu", "kamis", "jumat"]:
    harga_tiket -= 5000

if hari in ["sabtu", "minggu"] :
  harga_tiket += 10000

harga_tiket *= 1 - diskon/100
print(f'Harga tiket = {int(harga_tiket)}')
```

####Penjelasan

Kode tersebut digunakan untuk menghitung harga tiket berdasarkan umur pengunjung dan hari pembelian. Program meminta input umur dan hari, lalu menetapkan harga dasar tiket sebesar 50.000 rupiah dengan nilai diskon awal 0 persen. Jika umur kurang dari 12 tahun atau lebih dari 60 tahun, maka pengunjung mendapatkan diskon sebesar 50 persen. Jika umur berada di antara 12 sampai 60 tahun, tidak ada diskon umur, tetapi jika pembelian dilakukan pada hari kerja (Senin sampai Jumat), harga tiket dikurangi 5.000 rupiah. Setelah itu, jika pembelian dilakukan pada hari Sabtu atau Minggu, harga tiket akan ditambah 10.000 rupiah. Terakhir, harga tiket dikalikan dengan 1 dikurangi diskon dibagi 100 untuk menerapkan potongan harga sesuai persentase diskon, kemudian hasil akhirnya ditampilkan sebagai harga tiket yang harus dibayar.

```python
# jumlah baris
n = 4

for i in range(1, n + 1):      # loop untuk baris
    for j in range(i):         # loop untuk kolom (jumlah bintang per baris)
        print("*", end=" ")
    print()                    # pindah ke baris baru
```

####Penjelasan

Kode tersebut menggunakan looping bersarang (nested loop) untuk menampilkan pola segitiga bintang. Variabel n = 4 menentukan jumlah baris yang akan dicetak. Loop pertama for i in range(1, n + 1) berfungsi untuk mengatur baris, dimulai dari 1 hingga 4. Di dalamnya terdapat loop kedua for j in range(i) yang mencetak bintang sebanyak nilai i, sehingga pada baris ke-1 tercetak 1 bintang, baris ke-2 tercetak 2 bintang, dan seterusnya. Perintah print("*", end=" ") digunakan agar bintang dicetak dalam satu baris dengan spasi tanpa pindah baris, sedangkan print() di luar loop dalam berfungsi untuk berpindah ke baris baru setelah satu baris selesai dicetak.

```python
n = 4

for i in range(1, n + 1):          # loop baris
    for j in range(n - i):         # loop untuk spasi
        print(" ", end=" ")
    for k in range(i):             # loop untuk bintang
        print("*", end=" ")
    print()
```

####Penjelasan

Kode tersebut menggunakan looping bersarang untuk membuat pola segitiga rata kanan. Variabel n menentukan jumlah baris. Loop pertama mengatur baris dari 1 sampai n. Di dalamnya terdapat dua loop: loop pertama mencetak spasi sebanyak n - i agar bintang bergeser ke kanan, kemudian loop kedua mencetak bintang sebanyak i sesuai nomor baris. Setelah spasi dan bintang pada satu baris selesai dicetak, print() digunakan untuk pindah ke baris berikutnya. Polanya adalah jumlah spasi berkurang setiap baris, sedangkan jumlah bintang bertambah satu setiap baris.
